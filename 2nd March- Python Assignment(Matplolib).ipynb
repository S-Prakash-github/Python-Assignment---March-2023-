#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It provides a way to create a grid of subplots in which each subplot can show a different plot or a different view of the same plot.

```python
# define the data
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# create a 2x2 subplot grid
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))
# plot the data on each subplot
axs[0, 0].plot(x1, y1,c = "red")
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2,c = "blue")
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3,c = "orange")
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4, c = "green")
axs[1, 1].set_title('Line 4')
# set the title for the entire plot
fig.suptitle('Four Lines of Data')
# display the plot
plt.show()
```

A bar plot, also known as a bar chart, is a type of plot that is commonly used to show comparisons between different categories or groups.Bar plots can be used to show a variety of information, such as:

- Comparing values of different categories or groups
- Showing the distribution of a categorical variable
- Displaying frequency counts of a categorical variable

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize = (4,2))
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar plot showing company's profit")
plt.barh(company , profit,color = ["#E86161","#CE1D1D", "#9C0000","#6B0000"] )
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200

Box plots are used to get idea of the five number summary and for identifying the outliers in the dataset.
The box extends from the first quartile (Q1) to the thirdquartile (Q3) of the data, with a line at the median.  Thewhiskers extend from the box by 1.5x the inter-quartile range(IQR).

```python
import matplotlib.pyplot as plt
import numpy as np
# Create the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
# Combine the data into a list
data = [box1, box2]
```

```python
# Create the box plot
plt.figure(figsize = (6,5))
fig, ax = plt.subplots()
ax.boxplot(data)
# Add title and axis labels
ax.set_title('Box plot of box1 and box2')
ax.set_xlabel('Box')
ax.set_ylabel('Value')
# Show the plot
plt.show()
```
