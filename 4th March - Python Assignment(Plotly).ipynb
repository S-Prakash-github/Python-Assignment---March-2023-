#### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
t = sns.load_dataset("titanic")
fig  = go.Figure()
fig.add_trace(go.Scatter(x = t.age, y = t.fare, mode ="markers"))
fig.update_layout(
    title="Distribution of Fare by Age",
    xaxis_title="AGE",
    yaxis_title="FARE",
)
```

#### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
import plotly.express as go
import seaborn as sns
import pandas as pd
data = sns.load_dataset("iris")
# create a dataframe with the data
df = pd.DataFrame(data)

fig = go.scatter_matrix(df , color = 'species')

# update the title of the plot
fig.update_layout(title='Scatter Matrix Plot of Iris Dataset')

# show the plot
fig.show()
```

#### Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot is a graphical representation of the distribution of a numerical dataset. It displays the probability density function of the data, along with a histogram of the data itself. Distplot is a useful tool for visualizing the shape of a distribution, identifying outliers and skewness, and comparing multiple datasets.

```python
import plotly.express as px
import numpy as np

# generate a random dataset
np.random.seed(123)
data = np.random.normal(size=1000)

# create a distplot using Plotly Express
fig = px.histogram(data, x=data, nbins=30, opacity=0.5)

fig.update_layout(
    title="Distribution of a Random Dataset",
    xaxis_title="Data Values",
    yaxis_title="Density",
    bargap=0.1,
)

fig.show()
```

```python
import plotly.figure_factory as ff
import numpy as np

# generate a random dataset
## seed generates the same random data points again and again
np.random.seed(123)
## normal with 
data = np.random.normal(size=1000)
hist_data = [data]
group_labels = ['distplot'] # name of the dataset

fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(
    title="Distribution of a Random Dataset",
    xaxis_title="Data Values",
    yaxis_title="Density",
    bargap=0.1,
)

fig.show()
```
