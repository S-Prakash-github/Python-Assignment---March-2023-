#### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of eachplot

###### Five plots that we can plot using seaborn library are
- Scatter plot: A scatter plot displays the relationship between two variables. It is useful for identifying patterns, clusters, and outliers in the data. It can be created using the scatterplot() function in Seaborn.

- Line plot: A line plot displays the trend of a variable over time or across categories. It is useful for visualizing changes in the data over a continuous variable or category. It can be created using the lineplot() function in Seaborn.

- Bar plot: A bar plot displays the distribution of a categorical variable. It is useful for comparing the values of a variable across categories. It can be created using the barplot() function in Seaborn.

- Pairplot: It gives us an instant idea of the relationship between the numerical features inside the dataset.
- Boxplot : It is very useful in identifying the outliers in the dataset. It gives us a view of the five number summary.

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 
Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
## loading the dataset
fmri = sns.load_dataset("fmri")
fmri.head(2)
```

```python
## plotting line plot
sns.lineplot(x = "timepoint", y = "signal", data = fmri,color = "green")
plt.xlabel("Signal")
plt.ylabel("Timepoint")
plt.title("Timeplot on timepoint vs signal")
plt.show()
```

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'
Note:  pclass, age, and fare are columns in the titanic dataset

```python
## loading the dataset
titanic = sns.load_dataset("titanic")

## plotting a boxplot
sns.boxplot(x = "pclass", y = "age" ,data = titanic,palette = "Paired")
plt.xlabel("Passenger class")
plt.ylabel("Passenger age ")
plt.title("Box plot of Passenger class vs age")
plt.show()
```

```python
## plotting a boxplot
sns.boxplot(x = "pclass", y = "fare" ,data = titanic,palette = "Paired")
plt.xlabel("Passenger class")
plt.ylabel("Passenger Fare ")
plt.title("Box plot of Passenger class vs Fare")
plt.show()
```

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.head(2)
```

```python
## plotting a histplot
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack",kde = True,palette = "Paired")
plt.xlabel("Price")
plt.ylabel("Cut ")
plt.title("Histogram of price of diamond with cut ")
plt.show()
```

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
```

```python
## creating a pair plot
sns.pairplot(iris , hue = "species" ,palette = "Paired")
plt.show()
```

#### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
f = sns.load_dataset("flights")
```

```python
f.head(2)
```

```python
## plotting a heatmap
flights_data = f.pivot("month", "year", "passengers")
sns.heatmap(flights_data, cmap="YlGnBu")
plt.title("Heat map on monnth , year, passengers from the flight dataset ")
plt.show()
```
